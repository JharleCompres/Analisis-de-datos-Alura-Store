#1. Análisis de facturación



#### **Ingreso total de las tiendas: para conocer este monto sumaremos todas los valores de la columna precio de cada tienda.**

```python
try:
    # Data frame con la suma total de la columna precio de cada tienda
    df_total_ventas_tiendas = [
        tienda1["Precio"].sum(),
        tienda2["Precio"].sum(),
        tienda3["Precio"].sum(),
        tienda4["Precio"].sum()
    ]

    # Bucle que itera sobre el data frame con los datos
    for n, i in enumerate(df_total_ventas_tiendas, start=1):
        # Muestra el valor de i en cada interacción
        print(f"La suma del monto total de ventas de la tienda {n} es de: {i:.2f} MM")

except ValueError:
    print("Error: valores incorrectos")

except TypeError:
    print("Error: tipo de dato incorrecto")

except NameError:
    print("Error: variable no definida")

except SyntaxError:
    print("Error: sintaxis incorrecta")
```

###### **Nota:** Luego de realizar el cálculo observamos los ingresos totales de cada tienda, claramente se aprecia la diferencia entre cada una de ellas, siendo la tienda #1 con mayor ingreso y la tienda #4 la que menor ingreso ha obtenido.



#### **Bar chart: Ventas totales de las tiendas**

```python
try:
    # Importación de módulos
    from matplotlib import pyplot as plt
    from matplotlib import ticker

    # Validaciones básicas
    if 'df_total_ventas_tiendas' not in globals():
        raise NameError("La variable df_total_ventas_tiendas no está definida.")

    if not isinstance(df_total_ventas_tiendas, list):
        raise TypeError("df_total_ventas_tiendas debe ser una lista.")

    if len(df_total_ventas_tiendas) != 4:
        raise ValueError("df_total_ventas_tiendas debe contener 4 valores (una por cada tienda).")

    # Lista con los nombres de las tiendas
    tiendas = ["tienda1", "tienda2", "tienda3", "tienda4"]

    # Identificar las tiendas con mayor y menor ventas
    max_ventas = max(df_total_ventas_tiendas)
    min_ventas = min(df_total_ventas_tiendas)

    # Lista de colores: Resalta el total de ventas
    colores = []
    for i in df_total_ventas_tiendas:
        if i == max_ventas:
            colores.append("green")
        elif i == min_ventas:
            colores.append("red")
        else:
            colores.append("lightgray")

    # Crear gráfico
    plt.figure(figsize=(8, 5))
    plt.bar(tiendas, df_total_ventas_tiendas, color=colores)
    plt.title("Ventas totales por tienda", fontsize=14, weight="bold")
    plt.xlabel("Tiendas", weight="bold")
    plt.ylabel("Ventas totales (COP)", weight="bold")
    plt.grid(axis="y", linestyle="--", alpha=0.3)

    # Anotaciones
    plt.text(
        tiendas[df_total_ventas_tiendas.index(max_ventas)],
        max_ventas + 500,
        "Mayor venta",
        ha="center",
        color="green",
        weight="bold"
    )
    plt.text(
        tiendas[df_total_ventas_tiendas.index(min_ventas)],
        min_ventas + 500,
        "Menor venta",
        ha="center",
        color="red",
        weight="bold"
    )

    # Formatea eje Y a millones (MM)
    plt.gca().yaxis.set_major_formatter(
        ticker.FuncFormatter(lambda x, _: f'{x*1e-9:.1f} MM')
    )

    # Mostrar gráfico
    plt.tight_layout()
    plt.show()

except ValueError as e:
    print(f"Error de valor: {e}")

except TypeError as e:
    print(f"Error de tipo: {e}")

except NameError as e:
    print(f"Error de variable no definida: {e}")

except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

###### **Nota:** Podemos observar en el gráfico qué la tienda #1 destaca claramente en la mayor cantidad de ingresos en comparación con las demás tiendas, esto se puede deber a la ubicación o estrategia de ventas efectiva. En contraste tenemos la tienda #4 la cual muestra menor cantidad de ingresos, sería conveniente analizar si esto se debe a la ubicación, promoción o inventario limitado. Esto nos permite orientar los esfuerzos comerciales hacía donde el impacto seria mayor.


# 2. Ventas por categoría

#### **Cantidad de ventas agrupadas por categoría de cada una de las tiendas**

```python
try:
    # Dataframe que contiene los datos de cada tienda
    df_por_categoria = [tienda1, tienda2, tienda3, tienda4]

    # Lista que contiene los resultados de cada tienda
    ventas_por_categoria = []

    for c, d in enumerate(df_por_categoria):
        # Verifica que sea un DataFrame válido
        if d is None:
            raise ValueError(f"El dataframe de la tienda {c+1} está vacío o no fue definido.")

        # Agrupa por categorías y productos
        temp_df = (
            d.groupby("Categoría del Producto")["Producto"]
             .count()
             .reset_index(name="Total_ventas")
             .sort_values(by="Total_ventas", ascending=False)
        )

        # Agrega una columna al dataframe con el número de tienda
        temp_df["Tienda"] = c + 1

        # Agrega el resultado a la lista
        ventas_por_categoria.append(temp_df)

except KeyError as e:
    print(f"Error: La columna {e} no existe en uno de los dataframes.")

except ValueError as e:
    print(f"Error de valor: {e}")

except TypeError:
    print("Error de tipo: Verifique que los dataframes estén bien estructurados.")

except NameError:
    print("Error: Alguna variable no está definida (tienda, tienda2, etc).")

except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")

else:
    # Se ejecuta si no hubo errores
    for df in ventas_por_categoria:
        print(df.to_string(index=False))
```

### **Clustered bar chart: Ventas por categoría de cada tienda**

```python
try:
    import numpy as np

    # Validaciones
    if ventas_por_categoria is None:
        raise NameError("La variable ventas_por_categoria no está definida.")
    if not isinstance(ventas_por_categoria, list):
        raise TypeError("ventas_por_categoria debe ser una lista.")
    if len(ventas_por_categoria) != 4:
        raise ValueError("ventas_por_categoria debe contener 4 dataframes.")
    if any(df.empty for df in ventas_por_categoria):
        raise ValueError("Uno de los dataframes dentro de ventas_por_categoria está vacío.")

    # Unimos las ventas de las tiendas
    df_total = pd.concat(ventas_por_categoria, ignore_index=True)

    # Categorías y tiendas
    categorias = df_total["Categoría del Producto"].unique()
    tiendas = sorted(df_total["Tienda"].unique())

    x = np.arange(len(categorias))
    width = 0.15

    plt.figure(figsize=(14,7))

    for i, tienda_1 in enumerate(tiendas):
        subset = df_total[df_total["Tienda"] == tienda_1]
        subset = subset.set_index("Categoría del Producto").loc[categorias].reset_index()

        bars = plt.bar(
            x + (i * width),
            subset["Total_ventas"],
            width=width,
            label=f"Tienda {tienda_1}"
        )

        for bar in bars:
            height = bar.get_height()
            plt.text(
                bar.get_x() + bar.get_width()/2,
                height + 2,
                str(height),
                ha='center',
                va='bottom',
                fontsize=7,
                weight='bold'
            )

    plt.xticks(
        x + width * (len(tiendas) / 2 - 0.5),
        categorias,
        rotation=45,
        ha="right"
    )

    plt.title("Ventas por categoría de cada tienda", fontsize=14, weight="bold")
    plt.xlabel("Categoría del Producto", weight="bold")
    plt.ylabel("Total de ventas", weight="bold")
    plt.legend(title="Tiendas")
    plt.grid(axis="y", linestyle="--", alpha=0.3)

    plt.tight_layout()
    plt.show()

except ValueError as e:
    print(f"Error de valor: {e}")

except TypeError as e:
    print(f"Error de tipo: {e}")

except NameError as e:
    print(f"Error de variable no definida: {e}")

except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

###### **Nota:** En el gráfico visualizamos la comparación de cantidad de ventas en las categorias por cada tienda

### **Tienda con la menor venta en cada categoría**

```python
try:
  # Importamos módulo
  from itertools import count
  if df_total is None:
    raise NameError("La variable df_total no está definida.")
  if not isinstance(df_total, pd.DataFrame):
    raise TypeError("df_total debe ser un DataFrame.")
  if df_total.empty:
    raise ValueError("df_total está vacío.")
  if 'Categoría del Producto' not in df_total.columns:
    raise KeyError("La columna 'Categoría del Producto' no existe en df_total.")
  if 'Total_ventas' not in df_total.columns:
    raise KeyError("La columna 'Total_ventas' no existe en df_total.")
  if 'Tienda' not in df_total.columns:
    raise KeyError("La columna 'Tienda' no existe en df_total.")

  # Obtener el índice de la fila con menor venta por categoría
  idx_minimos = df_total.groupby("Categoría del Producto")["Total_ventas"].idxmin()

  # Extraer las filas completas
  tiendas_menor_venta = df_total.loc[idx_minimos].sort_values("Categoría del Producto").to_string(index=False)
except ValueError as e:
  print(f"Error de valor: {e}")
except TypeError as e:
  print(f"Error de tipo: {e}")
except NameError as e:
  print(f"Error de variable no definida: {e}")
except KeyError as e:
  print(f"Error de clave:{e}")
except Exception as e:
  print(f"Ocurrió un error inesperado: {e}")
else:
  # Muestra el resultado
  print(tiendas_menor_venta)
```

# 3. Calificación promedio de la tienda


#### **Tiendas organizadas de mayor calificación a menor calificación**

```python
try:
    # Construcción del DataFrame combinado
    df_calificacion = pd.concat([
        tienda1.assign(Tienda="Tienda 1"),
        tienda2.assign(Tienda="Tienda 2"),
        tienda3.assign(Tienda="Tienda 3"),
        tienda4.assign(Tienda="Tienda 4")
    ], ignore_index=True)

    # Verificación de columna requerida
    if "Calificación" not in df_calificacion.columns:
        raise KeyError("La columna 'Calificación' no se encuentra en el DataFrame combinado.")

    # Cálculo del promedio
    promedios = (
        df_calificacion.groupby("Tienda")["Calificación"]
        .mean()
        .round(2)
        .reset_index(name="Calificación Promedio")
        .sort_values(by="Calificación Promedio", ascending=False)
        .to_string(index=False)
    )

    # Imprimir resultado
    print(promedios)

except KeyError as e:
    print(f"ERROR: {e}")

except ValueError as e:
    print(f"ERROR al concatenar o procesar los DataFrames: {e}")

except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")

```

#### **Scatter Plot: Calificaciones de satisfacción de las tiendas**

```python
try:

  # Validaciones
  if df_calificacion is None:
    raise NameError("La variable df_calificacion no está definida.")
  if not isinstance(df_calificacion, pd.DataFrame):
    raise TypeError("df_calificacion debe ser un DataFrame.")
  if df_calificacion.empty:
    raise ValueError("df_calificacion está vacío.")
  if "Calificación" not in df_calificacion.columns:
    raise KeyError("La columna 'Calificación' no existe en df_calificacion")
  if "Tienda" not in df_calificacion.columns:
    raise KeyError("La columna 'Tienda' no existe en df_calificacion")
  if not all(isinstance(t, str) for t in df_calificacion["Tienda"]):
    raise TypeError("Los valores de la columna 'Tienda' no son cadenas de texto.")
  if not all(isinstance(c, (int, float)) for c in df_calificacion["Calificación"]):
    raise TypeError("Los valores de la columna 'Calificación' no son números.")

  # Data frame con las calificaciones promedio de cada tienda
  promedios_calificacion = df_calificacion.groupby(
      "Tienda")["Calificación"].mean().round(2).reset_index(
      name="Calificación Promedio").sort_values(by="Calificación Promedio",ascending=False)

  # Gráfico
  plt.figure(figsize=(8,5))
  plt.scatter(promedios_calificacion["Tienda"], promedios_calificacion["Calificación Promedio"], s=200)

  # Detalles del gráfico
  plt.title("Calificación promedio de la tienda", fontsize=14, weight="bold")
  plt.xlabel("Tiendas", weight="bold")
  plt.ylabel("Calificación Promedio", weight="bold")

  # Muestra el gráfico
  plt.show()
except ValueError as e:
  print(f"Error de valor: {e}")
except TypeError as e:
  print(f"Error de tipo: {e}")
except NameError as e:
  print(f"Error de variable no definida: {e}")
except KeyError as e:
  print(f"Error de clave: {e}")
except Exception as e:
  print(f"Ocurrió un error inesperado: {e}")
```

###### **Nota:** Se calculó la calificación promedio de satisfacción otorgada por los clientes a los productos de cada tienda. Esto permite evaluar objetivamente la calidad percibida y comparar el desempeño entre tiendas. Los resultados muestran que la **Tienda #4** presenta el nivel más alto de satisfacción, mientras que la **Tienda #1** requiere atención debido a un promedio inferior.

# 4. Productos más y menos vendidos

```python
from itertools import groupby
from numpy import size
try:
    df_productos = pd.concat([
        tienda1.assign(Tienda = "Tienda 1"),
        tienda2.assign(Tienda = "Tienda 2"),
        tienda3.assign(Tienda = "Tienda 3"),
        tienda4.assign(Tienda = "Tienda 4")
    ], ignore_index=True)

    if "Producto" not in df_productos.columns:
        raise KeyError("La columna de productos no se encuentra dentro del DataFrame")
    if "Tienda" not in df_productos.columns:
        raise KeyError("La columna de tiendas no se encuentra dentro del DataFrame")
    if "Precio" not in df_productos.columns:
        raise KeyError("La columna de precios no se encuentra dentro del DataFrame")
    if df_productos.empty:
        raise ValueError("El DataFrame está vacío")

    # Validar NaN
    if df_productos.isna().any().any():
        raise ValueError("El DataFrame contiene valores faltantes")

    # Validar tipos en la columna Tienda
    if not all(isinstance(t, str) for t in df_productos["Tienda"]):
        raise TypeError("Los valores de la columna 'Tienda' no son cadenas de texto")

    # Validar que los dataframes de origen no estén vacíos
    if any(df.empty for df in [tienda1, tienda2, tienda3, tienda4]):
        raise ValueError("Uno de los dataframes dentro de df_productos está vacío")

    # Obtiene la cantidad total de elementos
    ventas_productos = (
        df_productos.groupby(["Tienda","Producto"]).size().reset_index(
            name="Total_ventas")
    )

    # Muestra el top 5 de productos más vendidos en cada tienda
    top_5_mayor_venta = (
        ventas_productos.sort_values(["Tienda","Total_ventas"], ascending=[True, False]).groupby(
            "Tienda").head(5).reset_index(drop=True)
        )

    # Muestra el top 5 de productos menos vendidos en cada tienda
    top5_menos_vendidos = (
    ventas_productos.sort_values(["Tienda", "Total_ventas"], ascending=[True, True])
          .groupby("Tienda")
          .head(5)
          .reset_index(drop=True)
          )

    # Imprime los resultados
    print("\n📌 TOP 5 PRODUCTOS MÁS VENDIDOS POR TIENDA:\n")
    print(top_5_mayor_venta.to_string(index=False))

    print("\n📌 TOP 5 PRODUCTOS MENOS VENDIDOS POR TIENDA:\n")
    print(top5_menos_vendidos.to_string(index=False))

except ValueError as e:
    print(f"Error de valor: {e}")
except TypeError as e:
    print(f"Error de tipo: {e}")
except NameError as e:
    print(f"Error de variable no definida: {e}")
except KeyError as e:
    print(f"Error de clave: {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

###### **Nota:** Luego de haber calculado los productos con mayor y menor venta, realizamos un top 5 de productos más vendidos y top 5 productos menos vendidos organizados por tienda.

#### **Lollipop: Top 5 productos más vendidos por tienda**

```python
try:
    # Validaciones de los datos
    if top_5_mayor_venta is None:
      raise NameError("La variable top_5_mayor_venta no está definida.")
    if not isinstance(top_5_mayor_venta, pd.DataFrame):
      raise TypeError("top_5_mayor_venta debe ser un DataFrame.")
    if top_5_mayor_venta.empty:
      raise ValueError("top_5_mayor_venta está vacío.")
    if 'Tienda' not in top_5_mayor_venta.columns:
      raise KeyError("La columna 'Tienda' no existe en top_5_mayor_venta.")
    if 'Producto' not in top_5_mayor_venta.columns:
      raise KeyError("La columna 'Producto' no existe en top_5_mayor_venta.")
    if 'Total_ventas' not in top_5_mayor_venta.columns:
      raise KeyError("La columna 'Total_ventas' no existe en top_5_mayor_venta.")

    # funcion personalizada
    def lollipop_por_tienda(df, titulo):
        # Elementos unicos
        tiendas = df["Tienda"].unique()

        fig = plt.figure(figsize=(12, 8))

        # itera sobre cada subconjunto de los datos tomando los datos de cada tienda
        for i, tienda in enumerate(tiendas):
          subset = df[df["Tienda"] == tienda].sort_values("Total_ventas")
          productos = subset["Producto"]
          ventas = subset["Total_ventas"]

           # detalles del gráfico
          ax = fig.add_subplot(2, 2, i+1)
          ax.hlines(productos, 0, ventas)
          ax.plot(ventas, productos, "o")

          ax.set_title(tienda, fontsize=12)
          ax.set_xlabel("Ventas")

            # Título general BIEN UBICADO
        fig.suptitle(titulo, fontsize=16, y=1.02)

            # Ajusta los espacios entre los subplots
        fig.tight_layout(rect=[0, 0, 1, 0.95])

        plt.show()

        # TOP 5 MÁS VENDIDOS
    lollipop_por_tienda(top_5_mayor_venta, "Top 5 productos más vendidos por tienda")
except ValueError as e:
    print(f"Error de valor: {e}")
except TypeError as e:
    print(f"Error de tipo: {e}")
except NameError as e:
    print(f"Error de variable no definida: {e}")
except KeyError as e:
    print(f"Error de clave: {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

Por un lado, se identifican productos que dominan en ventas y que representan el punto fuerte de comercialización, reflejando una alta demanda y una estrategia que está funcionando bien.
