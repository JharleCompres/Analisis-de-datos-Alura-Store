#1. Análisis de facturación



#### **Ingreso total de las tiendas: para conocer este monto sumaremos todas los valores de la columna precio de cada tienda.**

```python
# Data frame con las suma total de la columna precio de cada tienda
df_total_ventas_tiendas = [tienda["Precio"].sum(),tienda2["Precio"].sum(),tienda3["Precio"].sum(),tienda4["Precio"].sum()]

# Bucle que itera sobre el data frame con los datos
for i in df_total_ventas_tiendas:
  # variable que accede al indice del valor de i y luego le suma 1 por cada interacción
  n = df_total_ventas_tiendas.index(i) + 1
  # Muestra el valor de i en cada interacción
  print(f"La suma del monto total de ventas de la tienda {n} es de: {i:.2f} MM")
```

###### **Nota:** Luego de realizar el cálculo observamos los ingresos totales de cada tienda, claramente se aprecia la diferencia entre cada una de ellas, siendo la tienda #1 con mayor ingreso y la tienda #4 la que menor ingreso ha obtenido.



#### **Gráfico de barras con ventas totales de las tiendas**

```python
# Importación de módulos de librería matplotlib
from matplotlib import pyplot as plt
from matplotlib import ticker

# Lista con los nombres de las tiendas
tiendas = ["tienda1","tienda2","tienda3","tienda4"]
# Data frame con datos de las tiendas
df_total_ventas_tiendas

# Identificar las tiendas con mayor y menor ventas
max_ventas = max(df_total_ventas_tiendas)
min_ventas = min(df_total_ventas_tiendas)

# Lista de colores: Resalta el total de ventas
colores = []
for i in df_total_ventas_tiendas:
  if i == max_ventas:
    colores.append("green")
  elif i == min_ventas:
    colores.append("red")
  else:
    colores.append("lightgray")

# Crea gráfico
plt.figure(figsize=(8,5))
plt.bar(tiendas,df_total_ventas_tiendas,color=colores)
plt.title("Ventas totales por tienda",fontsize=14,weight="bold")
plt.xlabel("Tiendas", weight="bold")
plt.ylabel("Ventas totales (COP)", weight="bold")
plt.grid(axis="y", linestyle="--", alpha=0.3)

# Anotaciones
plt.text(tiendas[df_total_ventas_tiendas.index(max_ventas)], max_ventas + 500, "Mayor venta", ha="center",color="green",weight="bold")
plt.text(tiendas[df_total_ventas_tiendas.index(min_ventas)], min_ventas + 500, "Menor venta", ha="center",color="red",weight="bold")

# formatea a MM la cifra
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x*1e-9:.1f} MM'))

# Muestra el gráfico
plt.tight_layout()
plt.show()
```

###### **Nota:** Podemos observar en el gráfico qué la tienda #1 destaca claramente en la mayor cantidad de ingresos en comparación con las demás tiendas, esto se puede deber a la ubicación o estrategia de ventas efectiva. En contraste tenemos la tienda #4 la cual muestra menor cantidad de ingresos, sería conveniente analizar si esto se debe a la ubicación, promoción o inventario limitado. Esto nos permite orientar los esfuerzos comerciales hacía donde el impacto seria mayor.


# 2. Ventas por categoría

#### **Cantidad de ventas agrupadas por categoría de cada una de las tiendas**

```python
#Dataframe que contiene los datos de cada tienda
df_por_categoria = [tienda,tienda2,tienda3,tienda4]

# Lista que contiene los resultados de cada tienda
ventas_por_categoria = []

for c,d in enumerate(df_por_categoria):
  # Agrupa por categorías y productos
  temp_df = d.groupby("Categoría del Producto")["Producto"].count().reset_index(name="Total_ventas").sort_values(by="Total_ventas",ascending=False)
  # Agrega una columna al dataframe con el numero de la tienda
  temp_df["Tienda"] = c + 1
  ventas_por_categoria.append(temp_df)
# Muestra el resultado
ventas_por_categoria
```

### **Gráfico con ventas por categoría de cada tienda**

```python
# Importamos las librerías
import numpy as np
import matplotlib.pyplot as plt

# Unimos las ventas de las tiendas
df_total = pd.concat(ventas_por_categoria, ignore_index=True)

# Categorías y tiendas
categorias = df_total["Categoría del Producto"].unique()
tiendas = sorted(df_total["Tienda"].unique())

# Posiciones base en X
x = np.arange(len(categorias))

# Ancho de cada barra
width = 0.15

plt.figure(figsize=(14,7))

for i, tienda in enumerate(tiendas):
    subset = df_total[df_total["Tienda"] == tienda]

    # Reordenar categorías para que coincidan con el eje X
    subset = subset.set_index("Categoría del Producto").loc[categorias].reset_index()

    # Dibujar barras
    bars = plt.bar(
        x + (i * width),
        subset["Total_ventas"],
        width=width,
        label=f"Tienda {tienda}"
    )

    # Agregar etiquetas encima de cada barra
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            height + 2,
            str(height),
            ha='center',
            va='bottom',
            fontsize=7,
            weight='bold'
        )

# Ajuste de etiquetas del eje X
plt.xticks(
    x + width * (len(tiendas) / 2 - 0.5),
    categorias,
    rotation=45,
    ha="right"
)

# Detalles agregados al gráfico
plt.title("Ventas por categoría de cada tienda", fontsize=14, weight="bold")
plt.xlabel("Categoría del Producto", weight="bold")
plt.ylabel("Total de ventas", weight="bold")
plt.legend(title="Tiendas")
plt.grid(axis="y", linestyle="--", alpha=0.3)

# Muestra el gráfico
plt.tight_layout()
plt.show()


```

###### **Nota** En el gráfico visualizamos la comparación de cantidad de ventas en las categorias por cada tienda

```python
# Importamos módulo
from itertools import count

# Obtener el índice de la fila con menor venta por categoría
idx_minimos = df_total.groupby("Categoría del Producto")["Total_ventas"].idxmin()

# Extraer las filas completas
tiendas_menor_venta = df_total.loc[idx_minimos].sort_values("Categoría del Producto")

# Muestra el resultado
print(tiendas_menor_venta)
```

###### **Nota:** Después de calcular cuáles tiendas obtuvieron las menores ventas por categoría, identificamos que la tienda **#2** presenta el rendimiento más bajo en **cuatro categorías**: Muebles, Deportes y diversión, Electrónicos y Juguetes.

Este comportamiento puede deberse a varios factores. Es posible que la estrategia comercial actual no esté logrando captar suficiente atención dentro de estas líneas de productos, o que la tienda necesite reforzar sus campañas de marketing. También podría influir el hecho de que los precios no se ajusten a la realidad económica del área donde opera la tienda. Otro punto a considerar es que estas categorías quizás no representan un interés significativo para la población local.

En cualquier caso, este hallazgo requiere un análisis más profundo para comprender por qué la tienda #2 muestra un rendimiento inferior en estas categorías en comparación con las demás. Con ello sería posible identificar oportunidades de mejora y tomar decisiones más informadas.
