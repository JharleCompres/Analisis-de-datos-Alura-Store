#### **Gráfico de barras con ventas totales de las tiendas**

```python
# Importación de módulos de librería matplotlib
from matplotlib import pyplot as plt
from matplotlib import ticker

# Listas con datos de las tiendas
tiendas = ["tienda1","tienda2","tienda3","tienda4"]
ventas = [tienda["Precio"].sum(),tienda2["Precio"].sum(),tienda3["Precio"].sum(),tienda4["Precio"].sum()]

# Identificar las tiendas con mayor y menor ventas
max_ventas = max(ventas)
min_ventas = min(ventas)

# Lista de colores: Resalta el total de ventas
colores = []
for i in ventas:
  if i == max_ventas:
    colores.append("green")
  elif i == min_ventas:
    colores.append("red")
  else:
    colores.append("lightgray")

# Crea gráfico
plt.figure(figsize=(8,5))
plt.bar(tiendas,ventas,color=colores)
plt.title("Ventas totales por tienda",fontsize=14,weight="bold")
plt.xlabel("Tiendas", weight="bold")
plt.ylabel("Ventas totales (COP)", weight="bold")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Anotaciones
plt.text(tiendas[ventas.index(max_ventas)], max_ventas + 500, "Mayor venta", ha="center",color="green",weight="bold")
plt.text(tiendas[ventas.index(min_ventas)], min_ventas + 500, "Menor venta", ha="center",color="red",weight="bold")

# formatea a MM la cifra
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x*1e-9:.1f} MM'))

# Muestra el gráfico
plt.show()
```

##### **OJO:** Podemos observar claramente en el gráfico qué la tienda #1 destaca claramente en la mayor cantidad de ventas en comparación con las demás tiendas, ésto sé puede deber a la ubicación o estrategia de ventas efectiva. En contraste tenemos la tienda #4 la cual muestra menor cantidad de ventas, sería conveniente analizar si ésto se debe a la ubicacion, promoción o inventario limitado. Ésto nos permite orientar los esfuerzos comeciales hacía donde el impacto seria mayor.
