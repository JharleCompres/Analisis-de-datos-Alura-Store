#### **Gráfico de barras con ventas totales de las tiendas**

```python
# Importación de módulos de librería matplotlib
from matplotlib import pyplot as plt
from matplotlib import ticker

# Listas con datos de las tiendas
tiendas = ["tienda1","tienda2","tienda3","tienda4"]
ventas = [tienda["Precio"].sum(),tienda2["Precio"].sum(),tienda3["Precio"].sum(),tienda4["Precio"].sum()]

# Identificar las tiendas con mayor y menor ventas
max_ventas = max(ventas)
min_ventas = min(ventas)

# Lista de colores: Resalta el total de ventas
colores = []
for i in ventas:
  if i == max_ventas:
    colores.append("green")
  elif i == min_ventas:
    colores.append("red")
  else:
    colores.append("lightgray")

# Crea gráfico
plt.figure(figsize=(8,5))
plt.bar(tiendas,ventas,color=colores)
plt.title("Ventas totales por tienda",fontsize=14,weight="bold")
plt.xlabel("Tiendas", weight="bold")
plt.ylabel("Ventas totales (COP)", weight="bold")
plt.grid(axis="y", linestyle="--", alpha=0.3)

# Anotaciones
plt.text(tiendas[ventas.index(max_ventas)], max_ventas + 500, "Mayor venta", ha="center",color="green",weight="bold")
plt.text(tiendas[ventas.index(min_ventas)], min_ventas + 500, "Menor venta", ha="center",color="red",weight="bold")

# formatea a MM la cifra
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x*1e-9:.1f} MM'))

# Muestra el gráfico
plt.tight_layout()
plt.show()
```

##### **OJO:** Podemos observar claramente en el gráfico qué la tienda #1 destaca claramente en la mayor cantidad de ventas en comparación con las demás tiendas, esto sé puede deber a la ubicación o estrategia de ventas efectiva. En contraste tenemos la tienda #4 la cual muestra menor cantidad de ventas, sería conveniente analizar si esto se debe a la ubicación, promoción o inventario limitado. Esto nos permite orientar los esfuerzos comerciales hacía donde el impacto seria mayor.


# 2. Ventas por categoría

**Tienda #1**

```python
ventas_categoria_tienda1 = tienda.groupby("Categoría del Producto")["Producto"].count().reset_index(name="Total_ventas").sort_values(by="Total_ventas",ascending=False)
ventas_categoria_tienda1
```

**Tienda #2**

```python
ventas_categoria_tienda2 = tienda2.groupby("Categoría del Producto")["Producto"].count().reset_index(name="Total_ventas").sort_values(by="Total_ventas",ascending=False)
ventas_categoria_tienda2
```

**Tienda #3**

```python
ventas_categoria_tienda3 = tienda3.groupby("Categoría del Producto")["Producto"].count().reset_index(name="Total_ventas").sort_values(by="Total_ventas",ascending=False)
ventas_categoria_tienda3
```

**Tienda #4**

```python
ventas_categoria_tienda4 = tienda4.groupby("Categoría del Producto")["Producto"].count().reset_index(name="Total_ventas").sort_values(by="Total_ventas",ascending=False)
ventas_categoria_tienda4
```

### **Gráfico con ventas por categoría de cada tienda**

```python
# Creamos columna y asignamos valor
ventas_categoria_tienda1["Tienda"] = "tienda1"
ventas_categoria_tienda2["Tienda"] = "tienda2"
ventas_categoria_tienda3["Tienda"] = "tienda3"
ventas_categoria_tienda4["Tienda"] = "tienda4"

# Unimos los valores de las tiendas en una lista
df_total = pd.concat([ventas_categoria_tienda1,ventas_categoria_tienda2,ventas_categoria_tienda3,ventas_categoria_tienda4])

# Asignamos el tamaño de la figura del gráfico
plt.figure(figsize=(10,6))
# Se toma el sub-conjunto por cada tienda
for Tiendas in df_total["Tienda"].unique():
  subset = df_total[df_total["Tienda"] == Tiendas]
  # Graficamos los resultados en barras
  plt.bar(subset["Categoría del Producto"], subset["Total_ventas"], label=Tiendas)
# Añadir al grafico
plt.title("Ventas por categoría de cada tienda",fontsize=14,weight="bold")
plt.xlabel("Categoría del Producto", weight="bold")
plt.ylabel("Total de ventas", weight="bold")
plt.legend(title="Tiendas")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.3)

# Mostrar el gráfico
plt.tight_layout()
plt.show()

```

#### **OJO:** Podemos observar en él siguiente gráfico qué la tienda #4 posee mayor estabilidad en las diferentes categorías, esto indica que tiene mayor operación estable y diversificación en ventas. Ha diferencia de las demás tiendas qué en categorías cómo: muebles, electrodomésticos, deportes y diversión, instrumentos musicales y libros. Presentan un mayor desempeño relativo lo que puede indicar posibles nichos o ventajas específicas.
